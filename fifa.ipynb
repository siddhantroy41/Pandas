```python
import pandas as pd 
fifa_df=pd.read_csv("2018 FIFA World Cup Squads.csv")
fifa_df
```

```python
#Group every team according to the data:
grp_a=fifa_df.loc[fifa_df["Group"]=="A","Team"].value_counts()
grp_b=fifa_df.loc[fifa_df["Group"]=="B","Team"].value_counts()
grp_c=fifa_df.loc[fifa_df["Group"]=="C","Team"].value_counts()
grp_d=fifa_df.loc[fifa_df["Group"]=="D","Team"].value_counts()
grp_e=fifa_df.loc[fifa_df["Group"]=="E","Team"].value_counts()
grp_f=fifa_df.loc[fifa_df["Group"]=="F","Team"].value_counts()
grp_g=fifa_df.loc[fifa_df["Group"]=="G","Team"].value_counts()
grp_h=fifa_df.loc[fifa_df["Group"]=="H","Team"].value_counts()
print("Group A :")
for team , p in grp_a.items():
    print(team)
print("Group B :")
for team , p in grp_b.items():
    print(team)
print("Group C :")
for team , p in grp_c.items():
    print(team)
print("Group D :")
for team , p in grp_d.items():
    print(team)
print("Group E :")
for team , p in grp_e.items():
    print(team)
print("Group F :")
for team , p in grp_f.items():
    print(team)
print("Group G :")
for team , p in grp_g.items():
    print(team)
print("Group H :")
for team , p in grp_h.items():
    print(team)
```
