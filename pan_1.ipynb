PANDAS TUTORIAL: First we have to import pandas library and give it an alias name

```python
import pandas as pd

df=pd.read_csv("Pokemon.csv")
df

#pd.read_csv reads the csv file and convert it into a tabular form also called "DATAFRAME" we will learn more about dataframe later.
```

SHAPE OF THE DATAFRAME:

```python
df.shape
#(800,13) is 800 rows and 13 columns
```

VIEWING THE DATAFRAME :

TO SEE EVERY ROW AND COLUMN WE CAN SET THE MAX_COLUMS AND MAX_ROWS TO THE MAX VALUE WE WANT 

```python
pd.set_option('display.max_columns',13)
pd.set_option('display.max_rows',800)
```

```python
df
#now it will show all the rows and all the columns
```

TO SEE SOME SPECIFIC NUMBER OF ROWS WE CAN USE HEAD AND TAIL METHOD

```python
df.head(10)
#it will show me the first ten records
```

```python
df.tail(10)
#it will show you the last ten rows or records of the dataframe
```

TO GET A BRIEF ABOUT YOUR DATA WE CAN USE INFO() METHOD

```python
df.info()
```

DESCIRBE METHOD SHOWS YOU THE DESCRIPTIVE STATS OF EACH COLUMNS

```python
df.describe()
```
