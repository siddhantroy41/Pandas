INDEXES: SET , RESET AND USE INDEXES

```python
import pandas as pd
people={
    "First" :["Siddhant","Rupaniwta","Karan"],
    "Last" :["Roy","Sahani","Yadav"],
    "Email":["siddhantroy@gmail.com","rupaniwtasahani@gmail.com","karanyadaav@gmail.com"]
}
df=pd.DataFrame(people)
df
```

SETTING CUSTOM INDEXES : YOU CAN SET ANY COLUMN AS YOUR INDEX BUT IT'S SMART TO USE THE COLUMN WHICH IS UNIQUE .

```python
#HERE WE WILL USE EMAIL AS CUSTOM INDEX AS IT IS UNIQUE FOR EVERY RECORD:
#it will not change the actual dataframe .
df.set_index("Email")
```

```python
#To make the above changes in the actual dataframe you have to set inpplace="True"
df.set_index("Email",inplace=True)
df

#now email has become the custom index for the actual dataframe.
```

```python
#Since each row have their specific indexes as their email we can access each row by using it's label or the custom index we just created .

df.loc["siddhantroy@gmail.com"]
```

RESETING THE INDEXES :

```python
df.reset_index(inplace=True)
df
```

```python
df_2=pd.read_csv("Pokemon.csv")
df_2

#Set the name column as the custom index:
new_data=df_2.set_index("Name")
new_data
```

```python
#Show the records of Bulbasaur
new_data.loc["Bulbasaur"]
```

```python
#Show the records from Bulbasaur to Charmander :

new_data.loc["Bulbasaur":"Charmander"]
```

```python
#Show records from Bulbasaur to Charmander but just for Type 1 and Type 2:

new_data.loc["Bulbasaur":"Charmander","Type 1":"Type 2"]
```

```python
#You can set the indexes in ascending and descending order:
#Ascending:
asc_data=df_2.sort_index()
asc_data
```

```python
#Descending :

des_data=df_2.sort_index(ascending=False)
des_data
```
