Filtering - Using Conditionals to Filter Rows and Columns

```python
import pandas as pd
people={
    "First" :["Siddhant","Rupaniwta","Karan"],
    "Last" :["Roy","Sahani","Yadav"],
    "Email":["siddhantroy@gmail.com","rupaniwtasahani@gmail.com","karanyadaav@gmail.com"]
}
df=pd.DataFrame(people)
df
```

Making a filter and then searching for that filter in that dataframe:

```python
filt=(df["Last"]=="Roy")
filt                #filt is a series of bool expressions with True and False if the condition matches it shows True otherwise False 

#Now we will search for that filter in the dataframe:
df[filt]
```

Another Way to filter data : Using loc() function , because you can also filter a specific column with that filter together :

```python
filt=(df["Last"]=="Roy")
df.loc[filt,"Email"]   #you can see that we used the filter to search for the row which have Roy as last name and then further we accessed the Email with it 
```

Filtering data using conditions (& and |)

```python
df_2=pd.read_csv("Pokemon.csv")
df_2

#Show the records where Attack is greater than 70 and Type 1 is Grass:

df_2.loc[(df_2["Attack"]>=70) & (df_2["Type 1"]=="Grass")]

#U can directly put the conditions inside the loc or use a variable to store the filter and then pass it inside the loc().
```

```python
#Using a variable to store the filter :
filt2= (df_2["Attack"]>=70) & (df_2["Type 1"]=="Grass")
df_2.loc[filt2]
```

```python
#If you have a filter but you want the results opposite to that filter,like above the conditions are Attack should be more than 70 and Type 1 should be Grass , if we use "~" symbol before the filter we will get the data opposite of the filter.
 
df_2.loc[~filt2]
```

```python
#Show the Type 1 and Type 2 of the pokemons which have Attack of more 70 and defence less than 50:
filt3=(df_2["Attack"]>=70) & (df_2["Defense"]<=50)
df_2.loc[filt3,"Type 1":"Type 2"]
```

```python
#Q.How many Pokémon are there from each generation?
generation_count= df_2["Generation"].value_counts().sort_index()
for gen,count in generation_count.items():
    print(f"Generation {gen} has {count} Pokémon.")
print("\n")
print(generation_count)   # it returns a series 
print(generation_count.items())      #it is an iterable object which we can access using two variables since it has a pair of values together 


#NOTE: value_counts returns a series and .items() makes the series into a iterable object of index value pair
```

```python
#Q.What are the most common primary and secondary types of Pokémon?

# Count the occurrences of each primary and secondary type
primary_type_counts = df_2['Type 1'].value_counts()
secondary_type_counts = df_2['Type 2'].value_counts()

common_type1=primary_type_counts.idxmax()
print(common_type1)
common_type2=secondary_type_counts.idxmax()
print(common_type2)



#NOTE: idmax() gives you the label of the value that have most occurences:
# suppose this is a series:Fire  50 , Water  80 , Normal  70
#so it will print water becoz it has the highest count 
```

```python
#q.Which Pokémon have the highest total stats, and how do their individual stats (Attack, Defense, etc.) compare?

highest_total=df_2["Total"].idxmax()
df_2.loc[highest_total]
```

```python
legendary=df_2["Legendary"]==True
non_legendary=df_2["Legendary"]==False
legendary_filter=df_2.loc[legendary,"Total"].items()
non_legendary_filter=df_2.loc[non_legendary,"Total"].items()
for (i , legendary_total),(j,non_legendary_total) in zip(legendary_filter,non_legendary_filter):
    if legendary_total<non_legendary_total:
        print(f"{df_2.loc[i,"Name"]} is a legendary pokemon that have low stats than non legendary pokemons")
```
